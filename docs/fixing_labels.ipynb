```python
# import relevant libraries
import numpy as np
import pandas as pd

# load in the data
artists = pd.read_csv('../data/modified_data/artist_profiles.csv')
tracks = pd.read_csv('../data/modified_data/all_tracks.csv')
```

```python
# load in all the genre data to give artists labels
country = pd.read_csv('../data/raw_data/genre_data/country.csv')
edm = pd.read_csv('../data/raw_data/genre_data/edm.csv')
funk = pd.read_csv('../data/raw_data/genre_data/funk.csv')
hip_hop = pd.read_csv('../data/raw_data/genre_data/hip_hop.csv')
jazz = pd.read_csv('../data/raw_data/genre_data/jazz.csv')
latin = pd.read_csv('../data/raw_data/genre_data/latin.csv')
pop = pd.read_csv('../data/raw_data/genre_data/pop.csv')
rap = pd.read_csv('../data/raw_data/genre_data/rap.csv')
rock = pd.read_csv('../data/raw_data/genre_data/rock.csv')
soul = pd.read_csv('../data/raw_data/genre_data/soul.csv')
```

```python
# remember -> we added six other artists
# britney spears, madonna = pop
# bob dylan, elvis, chuck berry, beach boys = rock

# only keep name and genre colummns
country = country[['name', 'genre']]
edm = edm[['name', 'genre']]
funk = funk[['name', 'genre']]
hip_hop = hip_hop[['name', 'genre']]
jazz = jazz[['name', 'genre']]
latin = latin[['name', 'genre']]
pop = pop[['name', 'genre']]
rap = rap[['name', 'genre']]
rock = rock[['name', 'genre']]
soul = soul[['name', 'genre']]

# combine all into genre_df 
genre_df = pd.concat([country, edm, funk, hip_hop, jazz, latin, pop, rap, rock, soul])

# create a df for the six extra artists
extra_artists = pd.DataFrame([
  {'name': 'Britney Spears', 'genre': 'pop'},
  {'name': 'Madonna', 'genre': 'pop'},
  {'name': 'Bob Dylan', 'genre': 'rock'},
  {'name': 'Elvis Presley', 'genre': 'rock'},
  {'name': 'Chuck Berry', 'genre': 'rock'},
  {'name': 'Beach Boys', 'genre': 'rock'}
])

# concatenate the two dataframes
genre_df = pd.concat([genre_df, extra_artists])

# what does this data look like? 
genre_df.head()
```

```python
print("Total Artists:", len(artists), "\ngenre_df Total Artists:", len(genre_df))
```

```python
# maximum number of duplicates in genre_df
genre_df['name'].value_counts().max()
```

```python
# find where the name column of genre_df = ac/dc
genre_df[genre_df['name'] == 'AC/DC']
```

```python
# remove duplicate indices 
genre_df = genre_df.drop_duplicates()
```

```python
# group by name and aggregate genres into a list
genre_agg = genre_df.groupby('name')['genre'].apply(list).reset_index()

# create separate columns for genre1, genre2, and genre3
genre_agg['genre1'] = genre_agg['genre'].apply(lambda x: x[0] if len(x) > 0 else None)
genre_agg['genre2'] = genre_agg['genre'].apply(lambda x: x[1] if len(x) > 1 else None)
genre_agg['genre3'] = genre_agg['genre'].apply(lambda x: x[2] if len(x) > 2 else None)

# drop the original column 'genre'
genre_agg = genre_agg.drop('genre', axis=1)

# find artists that have a value in genre2 or genre3 that is not 'none'
genre_agg[(genre_agg['genre2'].notnull()) | (genre_agg['genre3'].notnull())]
```

```python
# are there any rows that repeat names? 
genre_agg['name'].value_counts().max()
```

```python
# we deleted about 10 artists that had corrupted or no data

# rename name column to match the artists dataframe (artist_name)
genre_agg = genre_agg.rename(columns={'name': 'artist_name'})

# Perform the left join with the artists dataframe
artists_df = pd.merge(artists, genre_agg, on='artist_name', how='left')

artists_df.head()
```

```python
# unique values of time_signature
artists_df['time_signature'].value_counts()
```

```python
# due to the huge imbalance in the time_signature column, we will drop it
artists_df = artists_df.drop('time_signature', axis=1)

# last look at the data
artists_df.head(20)
```

```python
# save to csv
artists_df.to_csv('../data/clean_data/artists.csv', index=False)
```
